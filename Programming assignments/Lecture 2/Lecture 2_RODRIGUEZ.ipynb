```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import arff


data, meta = arff.loadarff('caesarian.csv.arff')
df = pd.DataFrame(data)
df = df.applymap(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
for column in df.columns:
    if df[column].dtype == 'object':
        df[column] = pd.to_numeric(df[column], errors='coerce')
df = df.dropna()
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Delivery number', hue='Caesarian', data=df, palette='Set1')
plt.title('Caesarian Rate by Delivery Number')
plt.xlabel('Delivery Number')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Delivery time', hue='Caesarian', data=df, palette='Set2')
plt.title('Caesarian Rate by Delivery Time')
plt.xlabel('Delivery Time')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Heart Problem', hue='Caesarian', data=df, palette='Set3')
plt.title('Heart Problems and Caesarian')
plt.xlabel('Heart Problem')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Blood of Pressure', hue='Caesarian', data=df, palette='coolwarm', alpha=0.6)
plt.title(' Likelihood causes for Caearian Age vs Blood Pressure')
plt.xlabel('Age')
plt.ylabel('Blood Pressure')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['Caesarian'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=90, labels=['Non-Caesarian', 'Caesarian'])
plt.title('Caesarian Cases')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
